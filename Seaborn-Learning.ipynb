# Line Plot

```python
sns.lineplot(x="distance", y="year", data=planets.head(20), color="red")

plt.show()
```

# Bar Plot

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
sns.barplot(x="year", y="passengers", data=flights, hue="month", palette="Accent")
plt.show()
```

# Scatter Plot 

```python
sns.scatterplot(x='tip', y='total_bill', data=tips)
```

```python
sns.scatterplot(x='tip', y='total_bill', data=tips, hue='day', size='size', palette='YlGnBu')
```

```python
sns.scatterplot(x="year", y="passengers", data=flights, hue="month", palette="Accent")
plt.show()
```

# Histogram

```python
sns.histplot(flights['passengers'], bins=10, palette="Accent", kde=True)

plt.title("Distribution of Passengers")
plt.xlabel("Number of Passengers")
plt.ylabel("Frequency")

plt.legend()

plt.show()
```

# Box Plot

```python
sns.boxplot(x='day', y='tip', data=tips, hue='sex')

plt.show()
```

# Count Plot

```python
flights.head()
```

```python
sns.countplot(x='passengers', data=flights.head(20), hue="year")

plt.show()
```

# Heatmap
**A plot that displays data as a colored grid, where the color represents the value of the data.**

```python
# heatmap only works on numerical data and works on 2d array 

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Original 1D data
data = [2,3,4,5,6,7,8,8,9,3,2,1,122,34,33,6,56,76,89,99,87,66,54,32,90]

# Reshape the data into a 2D array (for example, into a 5x5 matrix)
reshaped_data = np.reshape(data, (5, 5))

# Create the heatmap
sns.heatmap(data=reshaped_data)

# Show the plot
plt.show()
```
